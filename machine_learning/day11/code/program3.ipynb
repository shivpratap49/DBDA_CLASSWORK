# clustering

## problem definition
- find the clusters within the dataset using DBScan

### pre-requisites

```python
import warnings

warnings.filterwarnings(action="ignore")
```

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### load the data

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

### pre-processing

```python
# remove unwanted columns
df.drop(['CustomerID', 'Gender', 'Age'], axis=1, inplace=True)
```

```python
# rename the columns
df.rename({
    "Annual Income (k$)": "Income", 
    "Spending Score (1-100)": "SpendingScore"
}, axis=1, inplace=True)
```

```python
df.head()
```

### clustering

```python
from sklearn.cluster import DBSCAN

# create an object
dbscan = DBSCAN(eps=12, min_samples=10)

# fit the data
dbscan.fit(df)
```

```python
dbscan.labels_
```

### result visualization

```python
plt.scatter(df['Income'][dbscan.labels_ == -1], df['SpendingScore'][dbscan.labels_ == -1])
plt.scatter(df['Income'][dbscan.labels_ == 0], df['SpendingScore'][dbscan.labels_ == 0])
plt.scatter(df['Income'][dbscan.labels_ == 1], df['SpendingScore'][dbscan.labels_ == 1])
plt.scatter(df['Income'][dbscan.labels_ == 2], df['SpendingScore'][dbscan.labels_ == 2])
plt.scatter(df['Income'][dbscan.labels_ == 3], df['SpendingScore'][dbscan.labels_ == 3])
plt.scatter(df['Income'][dbscan.labels_ == 4], df['SpendingScore'][dbscan.labels_ == 4])
```

### create a new data set

```python
from sklearn.datasets import make_moons

x, y = make_moons(n_samples=300, noise=0.03, random_state=123456)
```

```python
all_x = []
all_y = []
for index in range(300):
    all_x.append(x[index][0])
    all_y.append(x[index][1])

# create arrays
all_x = np.array(all_x)
all_y = np.array(all_y)
```

```python
plt.scatter(all_x, all_y)
```

### clustering

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.3)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
plt.scatter(all_x[dbscan.labels_ == 0], all_y[dbscan.labels_ == 0])
plt.scatter(all_x[dbscan.labels_ == 1], all_y[dbscan.labels_ == 1])
plt.title("DBScan Clustering")
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
```

```python
plt.scatter(all_x[kmeans.labels_ == 0], all_y[kmeans.labels_ == 0])
plt.scatter(all_x[kmeans.labels_ == 1], all_y[kmeans.labels_ == 1])
plt.title("KMeans clustering")
```

```python
from sklearn.cluster import AgglomerativeClustering

agglomerative = AgglomerativeClustering(n_clusters=2)
agglomerative.fit(x)
```

```python
plt.scatter(all_x[agglomerative.labels_ == 0], all_y[agglomerative.labels_ == 0])
plt.scatter(all_x[agglomerative.labels_ == 1], all_y[agglomerative.labels_ == 1])
plt.title("Hierarchical Clustering")
```
