# Clustering

## problem definition 
- find relavent groups / clusters in the data using KMeans

### pre-requisites

```python
import warnings
warnings.filterwarnings(action='ignore')
```

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### load the data

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

### Pre-processing

```python
# remove unwanted columns
df.drop(['CustomerID', 'Gender', 'Age'], axis=1, inplace=True)
```

```python
# rename the features
df.rename({
    "Annual Income (k$)": "Income",
    "Spending Score (1-100)": "SpendingScore"
}, axis=1, inplace=True)
```

```python
df.head()
```

### clustering

```python
from sklearn.cluster import KMeans

# create an object
kmeans = KMeans(n_clusters=5, random_state=123456)

# fit the data
kmeans.fit(df)
```

```python
# get clusters
kmeans.labels_
```

```python
# get centroids
kmeans.cluster_centers_
```

### add a dependent variable in the dataset

```python
# add a new column with category
# category0: low income, high spending
# category1: mid income, mid spending
# category2: high income, high spending
# category3: high income, low spending
# category4: low income, low spending
cateogories = ["low income, high spending", "mid income, mid spending", 
               "high income, high spending", "high income, low spending", 
               "low income, low spending"]
df['category'] = kmeans.labels_
```

```python
df.head()
```

### train a model for classification

```python
# split the data into x and y 
x = df.drop('category', axis=1)
y = df['category']
```

```python
# split the data into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train, y_train)
```

### model evaluation

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_test)
y_true = y_test

accuracy = accuracy_score(y_true, y_pred)
accuracy
```

### result visualization

```python
colors = ["red", "green", "blue", "brown", "magenta"]

for cluster_index in range(5):
    plt.scatter(df['Income'][kmeans.labels_ == cluster_index],
                df['SpendingScore'][kmeans.labels_ == cluster_index],
               color=colors[cluster_index], label=f"cluster-{cluster_index + 1}")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Clustering")
plt.legend()
```

### Elbow method

```python
# collect all WSS for different clusters
wss = []

for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(df)
    wss.append(kmeans.inertia_)

# plot the chart
x = np.arange(1, 11)
plt.scatter(x, wss)
plt.plot(x, wss)
```
