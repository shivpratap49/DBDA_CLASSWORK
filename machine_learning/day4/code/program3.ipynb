### Pre-processing data

```python
from sklearn.preprocessing import StandardScaler

# create an object of standard scaler
scaler = StandardScaler()

# fit the data into the scaler object
# fit method requires a 2d collection (data frame)
scaler.fit(df)
```

```python
scaler.mean_
```

```python
# get the scaled data
# returns a 2d array
array_scaled = scaler.transform(df)
array_scaled

# array_scaled = scaler.fit_transform(df)
```

```python
sns.boxplot(array_scaled)
```

```python
# create a data frame from array
df_scaled = pd.DataFrame(array_scaled, columns=["Experience", "Salary"])
df_scaled
```
