### pre-processing

```python
df.columns
```

```python
# split the data into x and y
x = df[['Malic_Acid', 'Ash_Alcanity', 'Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols',
        'Proanthocyanins', 'Hue', 'OD280', 'Proline']]
y = df['Customer_Segment']
```

# split the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

### train the models

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

def create_model_logistic_regression():
    # create the model
    model = LogisticRegression(max_iter=1000)
    
    # train the model
    model.fit(x_train, y_train)

    return model
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

def create_model_naive_bayes():
    # create the model
    model = GaussianNB()
    
    # train the model
    model.fit(x_train, y_train)

    return model
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

def create_model_knn():
    # create the model
    model = KNeighborsClassifier(n_neighbors=5)
    
    # train the model
    model.fit(x_train, y_train)

    return model
```

#### SVM

```python
from sklearn.svm import SVC

def create_model_svm():
    # create the model
    model = SVC()
    
    # train the model
    model.fit(x_train, y_train)

    return model
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

def create_model_decision_tree():
    # create the model
    model = DecisionTreeClassifier()
    
    # train the model
    model.fit(x_train, y_train)

    return model
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

def create_model_random_forest():
    # create a model
    model = RandomForestClassifier(n_estimators=500)

    # train the model
    model.fit(x_train, y_train)

    return model
```

### model evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model):
    # get the predictions
    y_pred = model.predict(x_test)
    y_true = y_test

    # get the metrics
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='micro')
    recall = recall_score(y_true, y_pred, average='micro')
    f1 = f1_score(y_true, y_pred, average='micro')

    return accuracy, precision, recall, f1
```

### Performance chart

```python
# collection of models
models = [
    ("Logistic Regression", create_model_logistic_regression(), (0, 0)),
    ("Naive Bayes", create_model_naive_bayes(), (0, 1)),
    ("KNN", create_model_knn(), (1, 0)),
    ("SVM", create_model_svm(), (1, 1)),
    ("Decision Tree", create_model_decision_tree(), (2, 0)),
    ("Random Forest", create_model_random_forest(), (2, 1))
]
```

```python
# collect all model's performances
performance_chart_data = []
for item in models:
    # item[0] => model name
    # item[1] => model
    
    accuracy, precision, recall, f1 = evaluate_model(item[1])
    performance_chart_data.append([item[0], accuracy, precision, recall, f1])

performance_chart = pd.DataFrame(performance_chart_data, 
                     columns=["Model Name", "Accuracy", "Precision", "Recall", "F1"])
performance_chart
```

### result visualization

```python
# create sub plots with 3 rows and 2 columns
fig, axs = plt.subplots(3, 2)

# create classification visualization for all models
for model_name, model, position in models:
    # item[0] = model name
    # item[1] = model
    # item[2] = chart position

    # get the predictions
    y_pred = model.predict(x_test)
    y_true = y_test
    
    axs[position[0], position[1]].scatter(
        x_test['Malic_Acid'][y_pred == 1], 
        x_test['Ash_Alcanity'][y_pred == 1],
        color="red", label="Customer Segment 1")
    
    axs[position[0], position[1]].scatter(
            x_test['Malic_Acid'][y_pred == 2], 
            x_test['Ash_Alcanity'][y_pred == 2],
            color="green", label="Customer Segment 2")
    
    axs[position[0], position[1]].scatter(
            x_test['Malic_Acid'][y_pred == 3], 
            x_test['Ash_Alcanity'][y_pred == 3],
            color="blue", label="Customer Segment 3")

    
    axs[position[0], position[1]].set_title(model_name)
    axs[position[0], position[1]].set(xlabel="Malic Aci", ylabel="Ash Alcanity")
```

### creating a custom model
- this model is dependent on different learners

```python
import statistics

# def predict_values_using_custom_model(x_test):
final_predictions = []

# collect all the observations
data = []

# get x_test row by row
index = 0
for index, row in x_test.iterrows():

    # collect all the answers from every model
    answers = []
    
    # pass every row to every model
    for model_name, model, position in models:
        y_pred = model.predict([
            [
                row['Malic_Acid'],
                row['Ash_Alcanity'],
                row['Total_Phenols'],
                row['Flavanoids'],
                row['Nonflavanoid_Phenols'],
                row['Proanthocyanins'],
                row['Hue'],
                row['OD280'],
                row['Proline']
            ]
        ])
        
        # collect every prediction
        answers.append(y_pred[0])

    # get the mode of answers for final prediction
    final_prediction = statistics.mode(answers)

    # collect all final predictions
    final_predictions.append(final_prediction)
    
    # print(f"answers = {answers}, final_prediction = {final_prediction}")

    # add final predcition into the answers
    answers.append(final_prediction)

    # add the real answer into the answers
    answers.append(y_test[index])
    
    data.append(answers)
    index += 1

observation_df = pd.DataFrame(data, columns=['Logistic Regression', 'Naive Bayes', 'KNN', 'SVM', 'Decision Tree', 'Random Forest', 'Final Answer', 'Observed Answer'])
observation_df
```

### custom model evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# get the predictions
y_pred = final_predictions
y_true = y_test

# get the metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='micro')
recall = recall_score(y_true, y_pred, average='micro')
f1 = f1_score(y_true, y_pred, average='micro')

print(f"accuracy = {accuracy}, precision = {precision}, recall = {recall}, f1 = {f1}")
```
