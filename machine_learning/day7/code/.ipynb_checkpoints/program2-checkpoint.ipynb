# Classification

## problem definition
- predict whether the customer will leave the bank by understanding the given data

### pre-requisits

```python
import warnings
warnings.filterwarnings(action="ignore")
```

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### load the data

```python
df = pd.read_csv('Churn.csv')
df.head(2)
```

### exploratory data analysis

```python
df.info()
```

```python
df.describe()
```

### pre-processing

```python
# remove the unwanted columns
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# encode the geography column
df['Geography'] = LabelEncoder().fit_transform(df['Geography'])

# encode the gender column
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
```

```python
df.cov()
```

```python
df.corr()
```

```python
# from correlation analysis, only following columns are important
# Gender, Balance, Age and IsActiveMember
```

```python
# select only required columns
df = df[['Gender', 'Age', 'Balance', 'IsActiveMember', 'Exited']]
```

```python
# check if the data is balanced
df['Exited'].value_counts()
```

```python
df.head()
```

```python
# balance the data
from imblearn.over_sampling import SMOTE

# create a smote object
smote = SMOTE(random_state=123456)

# split the data into x and y
x = df.drop('Exited', axis=1)
y = df['Exited']

# oversample the values
x_oversampled, y_oversampled = smote.fit_resample(x, y)
```

```python
# split the data into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
        x_oversampled, y_oversampled, train_size=0.7, random_state=123456)
```

### train the model

```python
from sklearn.linear_model import LogisticRegression

# create the model
model = LogisticRegression(max_iter=1000, C=0.01, penalty='l2', solver='lbfgs')

# train the model
model.fit(x_train, y_train)
```

### model evaulation

```python
y_pred = model.predict(x_test)
y_true = y_test
```

#### accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true, y_pred)
accuracy
```

#### precision

```python
from sklearn.metrics import precision_score
precision = precision_score(y_true, y_pred)
precision
```

### recall

```python
from sklearn.metrics import recall_score
recall = recall_score(y_true, y_pred)
recall
```

#### f1 score

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_true, y_pred)
f1
```

#### classification report

```python
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred))
```

#### ROC

```python
from sklearn.metrics import roc_auc_score
roc = roc_auc_score(y_true, y_pred)
roc
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_true, y_pred)

plt.plot(fpr, tpr)
```

### model optimization

```python
# this section will try to find the optimal values of hyper parameters of LogisticRegression
from sklearn.model_selection import GridSearchCV

# create a dictionary with all possible hyperparamters along with their possible values
hyper_prameters = {
    "penalty": ["l1", "l2", "elasticnet"],
    "C": [10, 5, 1.0, 0.1, 0.01, 0.001],
    "solver": ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'}]
}

# create a grid search object
grid_search_cv = GridSearchCV(estimator=model, param_grid=hyper_prameters)
```

```python
grid_search_cv
```

```python
# do the permutations and combinations of hyper parameters and their values
grid_search_cv.fit(x_train, y_train)
```

```python
grid_search_cv.best_score_
```

```python
grid_search_cv.best_params_
```

### result visualization

```python
plt.scatter(
    x_test['Age'][y_pred == 0], 
    x_test['Balance'][y_pred == 0], 
    color="green",
    label="Customers wont leave")

plt.scatter(
    x_test['Age'][y_pred == 1], 
    x_test['Balance'][y_pred == 1], 
    color="red",
    label="Customers will leave")

plt.legend()
plt.xlabel("Age")
plt.ylabel("Balance")
```

```python
plt.scatter(
    x_test['Age'][y_true == 0], 
    x_test['Balance'][y_true == 0], 
    color="green",
    label="Customers wont leave")

plt.scatter(
    x_test['Age'][y_true == 1], 
    x_test['Balance'][y_true == 1], 
    color="red",
    label="Customers will leave")

plt.legend()
plt.xlabel("Age")
plt.ylabel("Balance")
```
