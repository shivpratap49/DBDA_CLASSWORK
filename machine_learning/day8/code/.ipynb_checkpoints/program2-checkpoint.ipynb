# Classification

## problem definition
- predict species of a flower based on sepal and petal dimensions

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### load the data

```python
df = pd.read_csv('iris.csv')
df.head(3)
```

### exploratory data analysis

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['species'].value_counts()
```

### pre-processing

```python
from sklearn.preprocessing import LabelEncoder

# create encoder object
encoder = LabelEncoder()

# encode the species column
df['species'] = encoder.fit_transform(df['species'])

encoder.classes_
```

```python
df.cov()
```

```python
df.corr()
```

```python
# split the data into x and y
x = df.drop('species', axis=1)
y = df['species']
```

```python
# split the data into train and test sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

### train the models

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

# create the model
model = KNeighborsClassifier(n_neighbors=5)

# load the data into the model
model.fit(x_train, y_train)
```

#### logistic regression

```python
from sklearn.linear_model import LogisticRegression

# create the model
model = LogisticRegression(max_iter=1000)

# train the model
model.fit(x_train, y_train)
```

#### naive bayes

```python
from sklearn.naive_bayes import GaussianNB

# create the model
model = GaussianNB()

# train the model
model.fit(x_train, y_train)
```

### evaluate the model

```python
y_pred = model.predict(x_test)
y_true = y_test
```

#### confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true, y_pred)
cm
```

#### accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true, y_pred)
accuracy
```

#### precision

```python
from sklearn.metrics import precision_score
precision = precision_score(y_true, y_pred, average='micro')
precision
```

#### recall

```python
from sklearn.metrics import recall_score
recall = recall_score(y_true, y_pred, average='micro')
recall
```

#### f1 score

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_true, y_pred, average='micro')
f1
```

### result visualization

```python
plt.scatter(
        x_test['sepal_length'][y_test == 0], 
        x_test['sepal_width'][y_test == 0], 
        color="green",
        label="setosa")

plt.scatter(
        x_test['sepal_length'][y_test == 1], 
        x_test['sepal_width'][y_test == 1], 
        color="red",
        label="versicolor")

plt.scatter(
        x_test['sepal_length'][y_test == 2], 
        x_test['sepal_width'][y_test == 2], 
        color="blue",
        label="verginica")


plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('observed values')
```

```python
plt.scatter(
        x_test['sepal_length'][y_pred == 0], 
        x_test['sepal_width'][y_pred == 0], 
        color="green",
        label="setosa")

plt.scatter(
        x_test['sepal_length'][y_pred == 1], 
        x_test['sepal_width'][y_pred == 1], 
        color="red",
        label="versicolor")

plt.scatter(
        x_test['sepal_length'][y_pred == 2], 
        x_test['sepal_width'][y_pred == 2], 
        color="blue",
        label="verginica")


plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('KNN')
```
