# Clustering 

## problem definition
- divide the dataset into optimum number of clusters

### pre-requisites

```python
import warnings

warnings.filterwarnings(action="ignore")
```

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### load the data

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

### EDA

```python
df.info()
```

### pre-processing

```python
# remove the unwanted features
df.drop(['CustomerID', 'Gender', 'Age'], axis=1, inplace=True)
```

```python
# rename the features
df.rename({
    "Annual Income (k$)": "Income",
    "Spending Score (1-100)": "SpendingScore"
}, axis=1, inplace=True)
```

### cluster the data

```python
# decide the number of clusters
n_clusters = 5
```

```python
from sklearn.cluster import KMeans

# create an object of KMeans
kmeans = KMeans(n_clusters=n_clusters, random_state=123456)

# fit the data into kmeans object
kmeans.fit(df)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

### result visualization

```python
colors = ["red", "green", "blue", "brown", "black", "yellow"]

for cluster in range(n_clusters):
    plt.scatter(df['Income'][kmeans.labels_ == cluster], 
                df['SpendingScore'][kmeans.labels_ == cluster], 
                color=colors[cluster])
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Clusters")
```

### optimization

```python
from sklearn.cluster import KMeans

# collect wss for all the clusters
wss = []

for cluster in range(1, 10):
    # create kmeans with number of clsuters
    kmeans = KMeans(n_clusters=cluster, random_state=123456)

    # fit the data
    kmeans.fit(df)

    # collect the sum of squares
    wss.append(kmeans.inertia_)
```

```python
# create the number of clusters
clusters = np.arange(1, 10)

# create the eblow chart
plt.scatter(clusters, wss, label="WSS")
plt.plot(clusters, wss)

# hilight the eblow point
plt.scatter(5, wss[5], color="red", label="Elbow point")

plt.xlabel("clsuters")
plt.ylabel("WSS")
plt.title("Eblow Method")
plt.legend()
```
